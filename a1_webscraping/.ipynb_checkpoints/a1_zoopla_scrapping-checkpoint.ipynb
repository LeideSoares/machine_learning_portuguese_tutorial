```python
# importar bibliotecas
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.zoopla.co.uk/to-rent/property/south-london/?price_frequency=per_month&q=South%20London&results_sort=newest_listings&search_source=to-rent"
```

```python
resposta = requests.get(URL)
```

```python
resposta.status_code
```

```python
resposta.text
```

```python
BeautifulSoup(resposta.text, 'html.parser')
```

```python
sopa = BeautifulSoup(resposta.text, 'html.parser')
```

```python
sopa.find_all(class_='srp clearfix')
```

```python
for casa in  sopa.find_all(class_='srp clearfix'):
    preco = casa.find(class_="listing-results-price text-price").text.replace('\n', '').strip()
    contacto_agente =  casa.find(class_='agent_phone').text.replace('\n', '').strip()
    compartimento =  casa.find(class_='listing-results-attr').text.replace('\n', '').strip()
    print('Preco:', preco)
    print('Contacto do agente:', contacto_agente)
    print('Compartimento:', compartimento)
```

```python
casas_zoopla = []

for casa in  sopa.find_all(class_='srp clearfix'):
    preco = casa.find(class_="listing-results-price text-price").text.replace('\n', '').strip()
    contacto_agente =  casa.find(class_='agent_phone').text.replace('\n', '').strip()
    compartimento =  casa.find(class_='listing-results-attr').text.replace('\n', '').strip()
    
    casas_zoopla.append([preco, contacto_agente, compartimento])
```

```python
import pandas as pd
```

```python
pd.DataFrame(casas_zoopla, columns=['preco', 'contacto do agnete', 'compartimento']).to_csv('Casas da Zoopla.csv')
```
