## Importar biblioteca relevantes

```python
import requests
from bs4 import BeautifulSoup

# Para converter info am dados estruturados
import pandas as pd
```

```python
URL = "https://pt.indeed.com/jobs?q=engenheiro%20civil&l=Lisboa"
```

```python
resposta = requests.get(URL)
```

```python
resposta.status_code
```

```python
resposta.text
```

```python
BeautifulSoup(resposta.text, 'html.parser')
```

```python
sopa = BeautifulSoup(resposta.text, 'html.parser')
```

#### Lista de oferta

```python
lst_oferta = sopa.find_all(class_='jobsearch-SerpJobCard unifiedRow row result')
```

```python
i = 1
for oferta in lst_oferta:
    titulo = oferta.find(class_='title').text.replace('\n', '')
    descricao = oferta.find(class_='summary').text
    try:
        empresa = oferta.find(class_='company').text.strip()
    except:
        empresa = 'Nao ha nome da empresa'
    
    i += 1 # --->  i = i + 1
    print('Titulo:', titulo)
    print('Empresa:', empresa)
```

### Converter em dados estrurados (DataFrame ou tabel rectangular)

```python
ofertas_indeed = []


for oferta in lst_oferta:
    titulo = oferta.find(class_='title').text.replace('\n', '')
    descricao = oferta.find(class_='summary').text.replace('\n', ' ')
    try:
        empresa = oferta.find(class_='company').text.strip()
    except:
        empresa = 'Nao ha nome da empresa'
    
    
    
    ofertas_indeed.append([titulo, empresa, descricao])
```

```python
indeed_trabalho_civil = pd.DataFrame(ofertas_indeed, columns=['Titulo', 'Empresa', 'Descricao'])
```
