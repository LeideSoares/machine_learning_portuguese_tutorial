```python
# PARA VISUALIZACAO DE DADOS
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd

# ficehiro para captura de dados - WORLDO METERS
from populacao_barometro_mundial import EvolucaoPopulacionalBM





# Nomes de Paises da CPLP (Comunidade de Paises da Lingua Oficial Portuguesa)
paises_cplp = [
                'Angola', 'Sao Tome & Principe', 'Cabo Verde', 'Mozambique', 
                'Guinea-Bissau', 'Portugal', 'Brazil'
            ]
```

## 2. Leitura de dados para extracao de urls (de paises da CPLP)
O dado a seguir  surge foi criado com codigo onde houve uma exposicao atravez de video.


```python
# Leitura de ficheiro com dados populacional de todos os paises de mundo e algumas dependencias.
paises_df  = pd.read_csv('../a1_webscraping/Population (2020).csv')
paises_df.set_index(['paises'], inplace=True)
paises_df.head()
```

## 3. Captura de captura para analise/visualizacao

```python
lista_de_paises = paises_df.loc[paises_cplp]
lista_de_paises
```

```python
urls_cplp =  lista_de_paises.url
```

### Captura de dados

```python
b_m = EvolucaoPopulacionalBM(urls_cplp)

evop_cplp = b_m.evo_populacao()

evop_cplp
```

# 4. Visualizacao

# 5. Analise
### Valores absoluto

```python
evop_cplp.plot(figsize = (15, 8))
```

## Valores relativos

```python
def racio(df):
    return df / df['1955-01-01']

evop_racio = evop_cplp.copy()

for pais in evop_cplp.columns:
    evop_racio[pais] = racio(evop_racio[pais])
```

```python
evop_racio
```

```python
evop_racio.mean()
```

```python

evop_racio.plot(figsize = (15, 8))
```

## Dados normalisado escala [0, 1]

```python
from sklearn.preprocessing import MinMaxScaler
escala = MinMaxScaler()
```

```python
evop_escalda = escala.fit_transform(evop_cplp)
evop_escalda
```

```python
evop_escalda = pd.DataFrame(evop_escalda, columns = evop_cplp.columns, index = evop_cplp.index)
evop_escalda
```

```python
evop_escalda.plot(figsize = (15, 8))
```

```python
evop_cplp['portugal'].plot(figsize = (15, 8))
```

# 6. CPLP Racio

```python
sns.set()
```

```python
plt.figure(figsize = (15, 8))
plt.plot(evop_racio, linewidth = 2);
plt.legend(evop_cplp.columns, fontsize=14)
```

```python
ax = evop_racio.plot(figsize = (15, 8))

for line, name in zip(ax.lines, evop_racio.columns):
    y = line.get_ydata()[-1]
    ax.annotate(name.title(), xy=(1,y), xytext=(6,0), color=line.get_color(), 
                xycoords = ax.get_yaxis_transform(), textcoords="offset points",
                size=16, va="center")
plt.xlim(['1955', '2020'])
plt.legend('')
plt.xlabel('Ano', fontsize = 16)
plt.ylabel('Proporcao de Crescimento', fontsize= 16)
plt.title('CPLP: Proporcao de Crescimento Anual [1955 - 2020] (dados: Worldometers.info)', fontsize = 20)
```

```python
ax = evop_cplp.plot(figsize = (15, 8), label = evop_cplp.columns)


leg = ax.legend()
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())

plt.title('CPLP: Crescimento Populacional [1955 - 2020] (dados: Worldometers.info)', fontsize = 20)

plt.xlim(['1955', '2020'])
plt.xlabel('Ano', fontsize = 16)
plt.ylabel('Total de Populacao', fontsize= 16);
plt.legend([col.title() for col in evop_cplp], fontsize = 16)
```

```python
ax = evop_escalda.plot(figsize = (15, 8))

leg = ax.legend()
for line, text in zip(leg.get_lines(), leg.get_texts()):
    text.set_color(line.get_color())


plt.title(f'CPLP: Normalisalcao da Evolucao de Populacional [1955 - 2020] (dados: Worldometers.info)', fontsize = 20)

plt.xlim(['1955', '2020'])
plt.xlabel('Ano', fontsize = 16)
plt.ylabel('Valor Normalisado\nMin(0), Max(1)', fontsize= 16)
plt.legend(fontsize = 16)
```
