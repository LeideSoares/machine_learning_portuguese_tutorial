A biblioteca oferece rapidez, flexibilidade, e um design expressivo para que os trabalhos com estrutura de dados sejam faceis e intuitivo. O nome pandas e uma conjugacao dos termos anglosaxonico "Panel" e "Data".

- etc.

## Leitura de dados (formato csv)

```python
df = pd.read_csv('../a1_webscraping/Population (2020).csv')
df
```

```python
df.set_index('paises',inplace=True)
```

```python
df.loc['Angola']['url']
```

```python
cplp_paises = [
                'Angola', 'Sao Tome & Principe', 'Cabo Verde', 'Mozambique', 
                'Guinea-Bissau', 'Portugal', 'Brazil'
            ]
```

```python
df.loc[cplp_paises]
```

```python
df.info()
```

```python
df.describe()
```

```python
df.select_dtypes('object').describe()
```

```python
df.regioes.unique()
```

```python
df.select_dtypes([float, int])
```

```python
df.groupby(['continentes'])[['Population (2020)']].sum()
```

```python
df.groupby(['continentes', 'regioes'])[['Population (2020)', 'Urban Pop %', 'World Share']].sum()
```

```python
def funcao_louca(valor):
    return sum(valor)/99999
```

```python
import numpy as np
```

```python
df.groupby(['continentes', 'regioes'])[['Population (2020)']].agg({'sum', 'mean', 'std', np.mean, funcao_louca})
```
