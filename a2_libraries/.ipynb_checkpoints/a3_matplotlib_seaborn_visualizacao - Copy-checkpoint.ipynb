## Valores relativos

```python
evop_racio  = evop_cplp.copy()

for pais in evop_cplp.columns:
    evop_racio[pais] = evop_cplp[pais] / evop_cplp[pais]['1955-01-01']
```

```python
evop_racio
```

```python
evop_racio.plot(figsize = (15, 8))
```
