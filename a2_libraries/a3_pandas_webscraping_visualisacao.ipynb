```python
df_cplp = pd.DataFrame()

for pais in cplp_paises:
    print(pais)
    df_cplp[pais] = paises_cplp[paises_cplp.Pais.str.contains(pais.lower().replace(' ', '-').replace('&', 'and'))]['Population']
```

```python
df_cplp['Year'] = pd.to_datetime(paises_cplp.Year.unique())
df_cplp.set_index('Year', inplace=True)
```

```python
df_cplp.sort_values('Year', inplace=True)
```

# Visualizacao de Evelucao Populacional

```python
df_cplp.plot( figsize=(16,8))
df_cplp['Sao Tome & Principe'].plot(linewidth=3, color ='r')
plt.title('Evolucao Populacional (absoluta) em CPLP [1955-2020]', fontsize=20)
plt.grid()
```

```python
def ratio_crescimento(valores):
    return valores / valores[0]


df_cplp_crescimento = pd.DataFrame()

for pais in df_cplp.columns:
    df_cplp_crescimento[pais] = df_cplp[pais].values / df_cplp[pais].values.min()
    
df_cplp_crescimento.index = df_cplp.index
```

```python
df_cplp_crescimento
```

```python
df_cplp_crescimento.plot( figsize=(16,8))
df_cplp_crescimento['Sao Tome & Principe'].plot(linewidth=3, color ='r')
plt.title('Crescimento Populacional dos Paises da CPLP [1955-2020]', fontsize=20)
plt.grid()
```
